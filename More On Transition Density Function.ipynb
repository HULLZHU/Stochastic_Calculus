## More on Transition Density Function

### 1.Definition of Transition Density Function

Given function $$P(W(s),s;W(t),t)\ \ \ \ (1)$$
Where
$$0 \leq s \leq T \ \ \ and \ \ 0 \leq t \leq T $$
and
$W(x)$ refers to the $state$ of a process at time $t$

Then we can defne $P(W(s),s;W(t),t)$ as the $transition$ $density$ $function$ by
$$Prob(a<W(t)<b\ at\ time\ t | at\ W(s)\ at\ time\ s)=\int^{b}_{a}P(W(s),s;W(t),t)dW(t)\ \ \ \ (2)$$

We can interpret it by following statement:

The probability that the variable $W(t)$ lies between $a$ and $b$ at time $t$ in the future, given that it started out with value $W(s)$ at time $s$.  

### 2.Forward Equation and Backward Equation
#### 2.1Forward Equation Derivation

Assume that at time $t$, the process reaches the value $y$, and we also assume that the process is subject to a trinomial tree, which is displayed as below.



It means for arriving value $y'$ at time $t'$, we can choose 3 forward status as followings:

$$(y'+\Delta y)\ at\ time\ (t'-\Delta t) \ with\ probability \ \alpha \ to\ reach\ \ \ \ (1.1)$$

$$y'\ at\ time\ (t'-\Delta t) \ with\ probability \ (1- 2\alpha) \ to\ reach\ \ \ \ (1.2)$$

$$(y'-\Delta y)\ at\ time\ (t'-\Delta t) \ with\ probability \ \alpha \ to\ reach\ \ \ \ (1.3)$$

thus, we can have following equation

$$P(y,t;y',t') = \alpha P(y,t;y'+\Delta y,t'-\Delta t) + (1-2\alpha)P(y,t;y',t'-\Delta t) + \alpha P(y,t;y'-\Delta y,t'-\Delta t)\ \ \ \ \ (*2)$$


Using $Taylor\ Expansion$
$$f(x_{0}+\Delta x)= f(x_{0})+f'(x_{0})\Delta x + \frac{1}{2!}f''(x_{0})\Delta x^{2}+ ……\ \ \ \ \ (3.1)$$

or

$$f(x_{0}+\Delta x, y_{0}+\Delta y) = f_{x}(x_0 + y_0)\Delta x + f_y(x_0,y_0)\Delta y + \frac{1}{2!}f_{xx}(x_0,y_0)\Delta x^2 + \frac{1}{2!}f_{yy}(x_{0},y_{0})\Delta y^2 + f_{xy}(x_0,y_0)\Delta x \Delta y + ……\ \ \ \ \ \ (3.2)$$

We can get that
$$P(y,t;y'+\Delta y, t'-\Delta t)\approx P(y,t;y',t')-\Delta t \frac{\partial P}{\partial t'}+\Delta y \frac{\partial P}{\partial t'} + \frac{1}{2}\Delta y^{2} \frac{\partial ^2 P}{\partial y'^2}+ …… (4.1)$$

$$P(y,t;y'-\Delta y, t' -\Delta t) \approx P(y,t;y',t')-\Delta t \frac{\partial P}{\partial t'}-\Delta y \frac{\partial P}{\partial y'}+\frac{1}{2}\Delta y^2 \frac{\partial ^2 P}{\partial y'^2}+……(4.2)$$

$$P(y,t;y',t'-\Delta t)\approx P(y,t;y',t')-\Delta t \frac{\partial p}{\partial t'} + …… (4.3)$$

Substitute (4.1),(4.2),(4.3) into (2), we get
$$P = \alpha(P-\Delta t \frac{\partial P}{\partial t'}+\Delta y \frac{\partial P}{\partial t'} + \frac{1}{2}\Delta y^{2} \frac{\partial ^2 P}{\partial y'^2}+ …)+(1-2\alpha)( P-\Delta t \frac{\partial P}{\partial t'}-\Delta y \frac{\partial P}{\partial y'}+\frac{1}{2}\Delta y^2 \frac{\partial ^2 P}{\partial y'^2}+…)+\alpha(P-\Delta t \frac{\partial p}{\partial t'} + …)\ \ \ \ \ (5.1)$$ 

We can simplify $(5.1)$ by cancelling terms, and we get

$$\Delta t \frac{\partial P}{\partial t'}=\alpha \Delta y^2 \frac{\partial ^2 P}{\partial y'^2}+… \ \ \ \ \ \ (*5.2)$$

or

$$\frac{\Delta P}{\Delta t'}=\alpha \frac{\Delta y^2}{\Delta t}\frac{\Delta ^2 P}{\Delta y'^2}+…\ \ \ \ \ \ (*5.3)$$

Note that
$$\displaystyle{\lim_{y \to 0, \Delta t \to 0}\alpha \frac{\Delta y^2}{\Delta t}}\ \ \ \ \  (6.1)$$
should converge to a $finite$ $limit$ to make the process be diffusive
In other words, $$O(\Delta y^2) \sim O(\Delta t)\ or\ O(\Delta y) \sim O(\sqrt{\Delta t})$$

Define 
$$\alpha \frac{\partial y^2}{\Delta t} = c^2 \ \ \ \ \ \ (*6.2)$$

Then $(*5.3)$ becomes

$$\frac{\partial P}{\partial t'} = c^2 \frac{\partial ^2 P}{\partial y'^2} \ \ \ \ (**7)$$

$(**7)$ is known as $Fokker-Planck\ Equation$ or $Forward\ Kolmogorov\ Equation$ (FKE)

#### 2.2 Intepretation for Forward Kolmogorov Equation:
From
$$\frac{\partial P}{\partial t'} = c^2 \frac{\partial ^2 P}{\partial y'^2}$$
We can find

(1)This is a PDE with 2 independent variable $y'$ and $t'$

(2)This is a $diffusion$ $equation$

(3)An example of Browian Motion

(4)quantity $c$ is related to volatility

### Solution of Trinomial Tree Kolmogrov Forward Equation
We have forward equation defined as
$$\frac{\partial P}{\partial t'}=c^2\frac{\partial^2P}{\partial y'^2}\ \ \ \ \ \ (1)Forward\ Equation$$

This is a special $diffusion$ function that we only need to use $combination$ of $y'$ and $t'$ as a variable to solve the equation

we assume the form of a solution is

$$P = t'^\alpha f(\frac{y'}{t'^b})\ \ \ \ \ (2)$$

where $a$, $b$ are constants and $f$ is the function of the variable  $\frac{y'}{t'^b}$ 

make
$$\xi = \frac{y'}{t'^b}\ \ \ \ \ \ (3)Replacement\ of\ Variable$$

now we want match $(1)$ by $(2)$,so we $only$ need to compute

$$\frac{\partial P}{\ t'} \ \ \ and \ \ \ \frac{\partial ^2 P}{\partial y'^2} \ \ \ \ (*target-functions)$$

$$\frac{\partial P}{\partial y'}= t'^\alpha \frac{f(\xi)}{\partial y'}=t'^\alpha \frac{\partial f(\xi)}{\partial \xi}\frac{\partial \xi}{\partial y'}$$

we get 

$$\frac{\partial P}{\partial y'}=t'^{a-b}\frac{df}{d\xi}$$

So we get

$$\frac{\partial^2P}{\partial y'^2} = t'^{a-b}\frac{d^2{f}}{d\xi ^2}\frac{d \xi}{d y'}$$

Finally we get

$$\frac{\partial^2P}{\partial y'^2} = t'^{a-2b}\frac{d^2f}{d\xi^2}\ \ \ \ \ (*4.target-function)$$

And we can get 
$$\frac{\partial P}{\partial t'} = at'^{a-1}f(\xi) - \frac{df}{d \xi}y't'^{a-b-1} \ \ \ \ \ (*5.target-function)$$

Now According to $(4)(5)$ and $(1)$,we make our PDE to be

$$at'^{^{a-1}}f(\xi)-by't'^{^{a-b-1}}\frac{df}{d\xi}=c^2t'^{a-2b}\frac{d^2f}{d\xi ^2}\ \ \ \ (6) From\ PDE\ to \ ODE$$

and then we get

$$af{\xi}-by't'^{^{-b}}\frac{df}{d\xi}=c^2t'^{^{1-2b}}\frac{d^2f}{d{\xi^2}}\ \ \ \ (7)$$

From $(3)$ we know $y' = t'^b\xi$, sub it into (7)

we have

$$af{\xi} -b\xi\frac{df}{d\xi}=c^{2}t'^{^{-2b+1}}\frac{d^2f}{d\xi^2} \ \ \ \ \ (8)$$

we need to make right hand of(8) is independent of t', so

$$b = \frac{1}{2}$$

then we make $(8)$ to be

$$af(\xi) -\frac{1}{2}\xi\frac{df}{d\xi}=c^2\frac{d^2f}{d\xi^2}\ \ \ \ \ (9)\ ODE$$

Again we need to discuss (2)

$$P = t'^\alpha f(\frac{y'}{t'^b})\ \ \ \ \ (2)$$

because (2) is a probability density function, thus we can have

$$\int_{-\infty}^{\infty}P(y',t')dy'=\int_{-\infty}^{\infty}t'^{^{a}}f(\frac{y'}{\sqrt{t'}})dy' = 1 \ \ \ \  (10)$$

make $\frac{y'}{\sqrt{t'}}=u$ or the same $y' = t'^{^\frac{1}{2}}u$

we have

$$t'^{^{a + 1/2}}\int_{-\infty}^{\infty}f(u)du=1\ \ \ \ \ (11)$$

so 

$$a = -\frac{1}{2} \ \ \ \ \ (12)$$

Now our ODE $(9)$ becomes

$$-\frac{1}{2}f(\xi)-\frac{1}{2}\xi\frac{df}{d\xi} = c^2\frac{d^2{f}}{d\xi^2}\ \ \ \ \ (13)$$

as the same as

$$-\frac{1}{2}\frac{d(\xi f(\xi))}{d\xi}=c^2\frac{d^2f}{d\xi^2} \ \ \ \ \ (14) Simplified\ Form$$

rewrite it as

$$c^2\frac{d(ln\ f(\xi))}{d\xi}=-\frac{\xi}{2} \ \ \ \ \ \ (15)$$

integrate it again, we have 

$$ln\ f(\xi)= -\frac{\xi ^2}{4 c^2} + c_1 \ \ \ \ \ \ (16)$$

Then we have

$$f(\xi)=Ae^{-\frac{\xi^2}{4c^2}} \ \ \ \ (17)$$

Now we know that $a=-\frac{1}{2}$ and $\int_{-\infty}^{\infty}P(y',t')dy'=\int_{-\infty}^{\infty}t'^{^{a}}f(\frac{y'}{\sqrt{t'}})dy' = 1=\int_{-\infty}^{\infty}\frac{1}{\sqrt{t'}}Ae^{-\frac{y'^{^2}}{4c^2t'}}dy'$,

thus we have

$$P = \frac{1}{2c\sqrt{\pi t'}}exp(-\frac{y'^{^2}}{4c^2t'}) \ \ \ \ \ \ (*18) Important Result$$

### Summary

By resolve the PDE
$$\frac{\partial P}{\partial t'}=c^2\frac{\partial^2P}{\partial y'^2}\ \ \ \ \ \ (1)Forward\ Equation$$

we find that

y' is normally distributed with $\mu=0$ and $\sigma^2 = 2c^2t'$


Generally Speaking, we can have following lemma for Trinomial Forward Equation

#### Lemma
If $y'$ is a value of a before mentioned trinomial random walk, and at time $t$, it has value $y$, then we can have its transition density function that

$$P(y,t;y',t') = \frac{1}{2c\sqrt{\pi(t'-t)}}exp(-\frac{(y'-y)^2}{4c^2(t'-t)})$$

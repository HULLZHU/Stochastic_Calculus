## $Stochastic\ Calculus\ and\ It\hat{o}'s Lemma$

### 1) Rule of Differentiation on Stochastic Variables

#### 1.1 Single Variable Case

For an ordinary variable, we have
$$f'(x) = \lim_{dx \to 0}\frac{f(x+dx)-f(x)}{dx} \ \ \ \ (1)$$

but for a stochastic variable, such as Browian Motion $W(t)$, we have

$$\frac{dW(t)}{dt} = \lim_{dt\to 0}\frac{dW(t)}{dt}$$

$$=\lim_{dt \to 0}\frac{O(\sqrt{dt})}{dt}$$

$$=\lim_{dt \to 0}\frac{1}{\sqrt{dt}} \ \ \ \ (*2)$$

so we can conclude that $NO$ limit value exists

### 2) Examples of SDEs

#### 2.1 GBM
$$\frac{dS}{S}=\mu dt + \sigma dX$$ 

where $X$ is a Brownian Motion 

switch off diffusion part: $\sigma = 0$

then we have

$$\int \frac{dS}{S} = \mu \int dt$$

Solve the ODE, we get

$$S = A e^{\mu t}$$

Make Initial Condition to be $A = S_0$, we get

$$S_t = S_0 e^{\mu t}$$

#### 2.2 Vasicek Model for Short Term Interest Rate
$$dr_t =(a - b r_t)dt + \sigma dX_t \ \ \ \ \ (8)$$

and then we have

$$dr_t = b(\frac{a}{b} - r_t)dt + \sigma dX_t \ \ \ \ \ (9)$$

Here $\frac{a}{b}$ is long run mean rate, and we can make $\frac{a}{b} = \hat{r}$

then we have

$$dr_t = -b(r_t - \hat{r})dt+\sigma dX \ \ \ \ (10)$$

if $r_t - \hat{r} > 0$, then $dr_t<0$  
if $r_t - \hat{r} < 0$, then $dr_t>0$  

#### 2.3 CIR Model for Short Term non-negative Interest Rate
$$dr_t = -\delta(r_t - \hat{r})dt + \sigma \sqrt{r_t}dX \ \ \ \ \ (11)$$

As $r_t \to 0$, we have
$$\sigma \sqrt{r_t} dX \to 0$$

So we can say that

$$\lim_{rt\to 0}Prob(dr_t <0) = 0 \ \ \ \ (12)$$

$$\int^{t}_{0}(dX)^2 = t \ \ \ \ \ \ (4)$$

### 3) Manipulate SDEs
#### 3.1 General Case
We assume that an options price $V$ is the function of stock price $S$, which is defined as $V(S)$

then we can have

$$V(S+dS)=\frac{\partial V}{\partial S}dS + \frac{1}{2}\frac{d^2V}{dS^2}dS^2 \ \ \ \ \ (18)$$

Where 

$$dS^2 = (\mu s dt + \sigma s dX)^2$$

$$=\mu^2s^2dt^2 +\sigma^2s^2dX^2 + 2\mu\sigma S^{2}dXdt$$

because

$dt^2 \to 0 $ and $dX^2 \to dt$ and $dXdt \to 0$

then we have

$$dS^2 = \sigma^2S^2dt \ \ \ \ (19)$$

sub it into $(18)$

we get

$$dV = (\mu S \frac{dV}{dS}+\frac{1}{2}\sigma^2S^2\frac{d^2V}{dS^2})dt +(\sigma S\frac{dV}{dS})dX \ \ \ \ \ (20)Final$$

#### 3.2 Special Case $V(S) = logS$

Now we assume

$$V(S) = logS \ \ \ \ \ (21)$$

differentiate it, we get

$$\frac{dV}{dS}=\frac{1}{S}\ \ \ \ \ \ (22)$$

and 

$$\frac{d^2V}{dS^2} = -\frac{1}{S^2} \ \ \ \ \ (23)$$

According to (20), then we get

$$dV = (\mu - \frac{1}{2}\sigma^2)dt + \sigma dX \ \ \ \ \ \ (24)$$

We can furthermore integrate $(24)$ between 0 and t

$$\int^{0}_{t}d(logS) = \int_{0}^{t}(\mu-\frac{1}{2}\sigma^2d\tau + \int^{t}_{0}\sigma dX $$

$$=(\mu -\frac{1}{2}\sigma^2)t + \sigma(X(t)-X(0)) \ \ \ \ \ (25)$$

which is equivalent to 

$$log(\frac{S(t)}{S(0)})=(\mu -\frac{1}{2}\sigma^2)t + \sigma(X(t)-X(0)) \ \ \ \ \ (26)$$

make $X_{0} = 0$, then we have

$$S(t) = S_{0}exp((\mu -\frac{1}{2}\sigma^2)t + \sigma X(t)) \ \ \ \ \ \ (27) Final$$



### 4) $It\hat{o}'s\ in\ Higher\ Dimension$

If the price of Option $V$ is the function of stock price $S$ and time $t$, 

$$V(S,t)$$

$$dV = \frac{\partial V}{\partial t}dt + \frac{\partial V}{\partial S}dS + \frac{1}{2}b^2\frac{\partial ^2V}{\partial S^2}dt \ \ \ \ \ (*28)$$

If we assume that 

$$dS= \mu S dt + \sigma S dX$$

we can get

$$dV = \frac{\partial V}{\partial t}dt + \frac{\partial V}{\partial S}dS + \frac{1}{2}\sigma ^ 2 S^2 \frac{\partial ^2 V}{\partial S^2}dt \ \ \ \ \ \ (* 29)$$

#### 4.2 Multi-Variable Case

$$V(S_1,S_2,t)$$

$$dS_1= a_1(S_1,S_2,t)dt + b_1(S_1,S_2,t) dX_1$$
